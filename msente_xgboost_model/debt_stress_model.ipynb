```python
categorical_features = ["employment_type"]

numerical_features = [
    column
    for column in X.columns
    if column not in categorical_features
]

print("Categorical features:")
print(categorical_features)

print("\nNumerical features:")
print(numerical_features)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "employment_encoder",
            OneHotEncoder(
                handle_unknown="ignore",
                sparse_output=False
            ),
            categorical_features
        )
    ],
    remainder="passthrough"
)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
print("Processed training shape:", X_train_processed.shape)
print("Processed testing shape:", X_test_processed.shape)
```

```python
feature_names = preprocessor.get_feature_names_out()

print(feature_names)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    objective="binary:logistic",
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(
    X_train_processed,
    y_train
)

print("Model training completed successfully.")
```

```python
y_probability = xgb_model.predict_proba(
    X_test_processed
)[:, 1]
y_probability
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

y_pred = xgb_model.predict(X_test_processed)

y_probability = xgb_model.predict_proba(
    X_test_processed
)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_probability)

print("MODEL PERFORMANCE")
print("-----------------")
print("Accuracy:", round(accuracy, 4))
print("Precision:", round(precision, 4))
print("Recall:", round(recall, 4))
print("F1 Score:", round(f1, 4))
print("ROC-AUC:", round(roc_auc, 4))

print("\nCONFUSION MATRIX")
print(confusion_matrix(y_test, y_pred))

print("\nCLASSIFICATION REPORT")
print(
    classification_report(
        y_test,
        y_pred,
        target_names=[
            "Safer Loan",
            "Debt Stress"
        ]
    )
)
```

```python
def predict_debt_stress(
    borrower_data,
    package
):
    """
    Predict whether a loan may cause debt stress.

    borrower_data must contain the 13 base input fields.
    """

    required_features = package[
        "base_input_features"
    ]

    missing_features = [
        feature
        for feature in required_features
        if feature not in borrower_data
    ]

    if missing_features:
        raise ValueError(
            f"Missing fields: {missing_features}"
        )

    prepared_data = borrower_data.copy()

    loan_amount = float(
        prepared_data["loan_amount"]
    )

    repayment_amount = float(
        prepared_data["repayment_amount"]
    )

    monthly_income = float(
        prepared_data["monthly_income"]
    )

    average_inflow = float(
        prepared_data["avg_monthly_inflow"]
    )

    if loan_amount <= 0:
        raise ValueError(
            "loan_amount must be greater than zero."
        )

    if monthly_income <= 0:
        raise ValueError(
            "monthly_income must be greater than zero."
        )

    if average_inflow <= 0:
        raise ValueError(
            "avg_monthly_inflow must be greater than zero."
        )

    # Automatically calculate derived features
    prepared_data["loan_cost_percentage"] = (
        (
            repayment_amount - loan_amount
        )
        / loan_amount
    ) * 100

    prepared_data[
        "repayment_to_income_ratio"
    ] = (
        repayment_amount / monthly_income
    )

    prepared_data[
        "repayment_to_inflow_ratio"
    ] = (
        repayment_amount / average_inflow
    )

    input_frame = pd.DataFrame(
        [prepared_data]
    )

    expected_columns = package[
        "model_input_features"
    ]

    input_frame = input_frame[
        expected_columns
    ]

    processed_input = package[
        "preprocessor"
    ].transform(input_frame)

    risk_probability = package[
        "model"
    ].predict_proba(
        processed_input
    )[0, 1]

    threshold = package["threshold"]

    debt_stress_prediction = int(
        risk_probability >= threshold
    )

    if risk_probability < 0.30:
        risk_level = "Low Risk"
        message = (
            "The loan appears relatively manageable "
            "based on the available information."
        )

    elif risk_probability < 0.40:
        risk_level = "Caution"
        message = (
            "Review the loan terms carefully "
            "before accepting."
        )

    elif risk_probability < 0.70:
        risk_level = "High Risk"
        message = (
            "This loan may create repayment "
            "difficulty."
        )

    else:
        risk_level = "Severe Debt Stress Risk"
        message = (
            "This loan may push the borrower "
            "into serious debt stress."
        )

    return {
        "risk_probability": round(
            float(risk_probability) * 100,
            2
        ),
        "risk_level": risk_level,
        "debt_stress_prediction":
            debt_stress_prediction,
        "prediction_label": (
            "Debt Stress"
            if debt_stress_prediction == 1
            else "Safer Loan"
        ),
        "message": message,
        "calculated_features": {
            "loan_cost_percentage": round(
                prepared_data[
                    "loan_cost_percentage"
                ],
                2
            ),
            "repayment_to_income_ratio": round(
                prepared_data[
                    "repayment_to_income_ratio"
                ],
                4
            ),
            "repayment_to_inflow_ratio": round(
                prepared_data[
                    "repayment_to_inflow_ratio"
                ],
                4
            )
        }
    }
```

```python
sample_borrower = {
    "age": 26,

    # 0=salaried, 1=self-employed,
    # 2=student, 3=freelancer, 4=unemployed
    "employment_type": 1,

    "monthly_income": 450000,
    "income_stability_score": 0.55,

    "avg_monthly_inflow": 520000,
    "avg_monthly_outflow": 430000,
    "avg_balance": 70000,

    "active_loans_count": 2,
    "missed_payments_6m": 1,

    "loan_amount": 200000,
    "repayment_amount": 300000,
    "loan_term_days": 14,

    # Zero means no collateral
    "collateral_value": 0
}

result = predict_debt_stress(
    sample_borrower,
    loaded_package
)

result
```
